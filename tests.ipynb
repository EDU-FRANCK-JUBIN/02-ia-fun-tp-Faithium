```python
import cv2
import sys
from matplotlib import pyplot as plt

imagePath = r'image0.jpg'
dirCascadeFiles = 'opencv-3.4/haarcascades_cuda/'
cascadefile = dirCascadeFiles + 'haarcascade_frontalface_default.xml'
classCascade = cv2.CascadeClassifier(cascadefile)
image = cv2.imread(imagePath)
plt.imshow(image)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(image)
```

```python
faces = classCascade.detectMultiScale(
    gray,
    scaleFactor=1.1,
    minNeighbors=5,
    minSize=(30, 30),
    flags=cv2.CASCADE_SCALE_IMAGE
)

print("Il y a {0} visage(s)".format(len(faces)))
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
plt.imshow(image)
```

```python
image2 = cv2.imread(imagePath)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cascadefile2 = dirCascadeFiles + 'haarcascade_frontalface_alt.xml'
classCascade2 = cv2.CascadeClassifier(cascadefile2)
faces2 = classCascade2.detectMultiScale(
    gray,
    scaleFactor=1.1,
    minNeighbors=5,
    minSize=(30, 30),
    flags=cv2.CASCADE_SCALE_IMAGE
)
print("Il y a {0} visage(s)".format(len(faces2)))
for (x, y, w, h) in faces2:
    cv2.rectangle(image2, (x, y), (x+w, y+h), (0, 255, 0), 2)
plt.imshow(image2)
```

```python
for (x, y, w, h) in faces2:
    crop_img=image2[y:y+h, x:x+w]
    plt.imshow(crop_img)
```
